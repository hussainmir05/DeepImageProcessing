<h1> Task 02<h1>


Read an image and resize it to dimension of 256 x 256. Next 
split the image into four equal parts and swap the diagonal 
parts of the image. Display both the input and output image.


```python
img=cv2.imread('/content/gray.jpeg',0)
cv2_imshow(img)

from google.colab.patches import cv2_imshow
print( 'size is equal to = 256x256')
resized = cv2.resize(img, (256,256), interpolation = cv2.INTER_AREA)
cv2_imshow(resized)

import matplotlib.pyplot as plt
(h, w) = img.shape[:2]
center_x,centre_y = (w // 2), (h // 2)
print( 'Four equal parts')

topleft = img[0:centre_y,0:center_x]
topright = img[0:centre_y,center_x:w]
bottomleft = img[centre_y:h,0:center_x]
bottomright = img[centre_y:h,center_x:w]

plt.subplot(2,2,1)
plt.imshow(topleft)
plt.subplot(2,2,2)
plt.imshow(topright)
plt.subplot(2,2,3)
plt.imshow(bottomleft)
plt.subplot(2,2,4)
plt.imshow(bottomright) 


```

```python
print('swap the diagonal parts of the image')
i=bottomright+bottomleft+topright+topleft 
cv2_imshow(i)
```
