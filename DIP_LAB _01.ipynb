<h3> Task 06<h3>

It is easy to convert a list to a set and vice versa. For
example, a list mylist = [1; 1; 2; 3; 4; 4; 5] can be converted
to a set using the command newset = set(mylist). The set can
be converted back to a list using newlist = list(newset).

```python
mylist = [1, 1, 2, 3, 4, 4, 5] 
print('mylist = ',mylist)

newset=set(mylist)
print('newset = ',newset)
newlist=list(newset)
print('newlist = ',newlist)
```

sets does allow a duplicate element while in lists there may be duplicate elements 

mylist has duplicate elements and when we convert it into set by set(mylist)there will be no duplicate elements and when we convert back to list there will be no repetation

Create a numpy array of size 5-by-5 containing all random
