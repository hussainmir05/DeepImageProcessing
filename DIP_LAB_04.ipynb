# task 04
Perform image masking to hide all other details in the image other than 
the face of subject.


```python
k="/content/ast.jpg"
image =cv2.imread(k)

mask = np.zeros(image.shape[:2], dtype="uint8")
cv2.rectangle(mask, (0, 90), (290, 450), 255, -1)
cv2_imshow(mask)

masked = cv2.bitwise_and(image, image, mask=mask)
cv2_imshow(masked)
cv2.waitKey(0)
```
